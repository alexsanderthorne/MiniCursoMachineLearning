Inicialmente, realizei a importação de bibliotecas ou módulos: o primeiro deles foi o dataset load_iris, da biblioteca sklearn — ou Scikit-learn, que é o local onde estão os dados que utilizaremos para criar nosso modelo. Logo em seguida, importei o modelo train_test_split, que irá dividir nosso dataset em dois conjuntos: treino e teste. Por fim, importei o algoritmo K-Nearest Neighbors, também conhecido como KNN. Também importei a biblioteca Matplotlib para criarmos gráficos, além de incluir o código %matplotlib inline, que fará com que nossos gráficos sejam gerado no próprio notebook e não em uma nova janela.

```python
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

Comecei o algoritmo criando uma variável iris, onde eu carrego o dataset. Após isso, eu exibo as chaves (ou colunas) que estão presentes no arquivo.

```python
iris = load_iris()
iris.keys()
```

Aqui criei quatro variáveis: X_train, X_test, y_train e y_test. Estas irão armazenar os dados de treino e teste que serão utilizadas por nosso modelo.

```python
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state = 0)
```

A melhor forma de inspecionarmos dados é visualizando-os. Para fazer isso, utilizamos a biblioteca Matplotlib e geramos um gráfico de dispersão

```python
fig, ax = plt.subplots(3,3, figsize=(15,15))

for i in range(3):
    for j in range(3):
        ax[i,j].scatter(X_train[:,j], X_train[:,i+1], c=y_train, s=60)
        
        if i == 2:
            ax[i,j].set_xlabel(iris['feature_names'][j])
        if j == 0:
            ax[i,j].set_ylabel(iris['feature_names'][i+1])
        if j > i:
            ax[i,j].set_visible(False)
```

Criei aqui a variável knn e atribui a ela o objeto KNeighborsClassifier. O argumento n_neighbors está indicando a quantidade de vizinhos que precisamos que nosso código leve em consideração antes de prever o tipo de flor. Neste caso, com n_neighbors=1, ele simplesmente irá verificar qual o ponto mais próximo. Se alterássemos esse valor para 3, por exemplo, ele cruzaria o ponto mais próximo entre três vizinhos antes de prevê-lo.

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

Na variável prediction, armazenei o resultado da função knn.predict, que irá prever qual tipo de flor para o conjunto X_test.

```python
prediction = knn.predict(X_test)
print('previsto: ', prediction)
print('esperado: ', y_test)
```

Por fim, mas não menos importante, devemos saber o quão preciso nosso modelo é.

```python
knn.score(X_test,y_test)
```
